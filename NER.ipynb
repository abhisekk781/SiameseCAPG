```python
#pip install en_core_web_sm
#pip install spacy
#Catastrophic forgetting issue
#save the updated nlp model
import en_core_web_sm
nlp = en_core_web_sm.load()
import pandas as pd
import numpy as np
import re
data = pd.read_excel('Software_names_data.xlsx')
with open('./Software_Names.txt', 'r') as f:
    techNames = list(set([" ".join(word.split('\n')[0].lower().split()) for word in f]))
df = data.dropna(subset = ['SoftWare_Names'])
sentences = list(set(df.sampleSample.values.tolist()))
```

```python
d = {'SAP' : 'techStack',
     'issue' : 'issueStack',
     'tenneco technologies' : 'techStack'}
```

```python
#tag = 'TECHNOLOGY STACK'
def getEntityTags(text, dictOfWordTag):
    array = []
    dict_ = {}
    for word, tag in arrayOfWordsTags.items():
        #Regex Matching
        l = r'\b{}\b'.format(word)
        if bool(re.search(l, text)):
            array.append((re.search(l, text).start(), re.search(l, text).end(), tag))
        else: continue
    if len(array)==0:
        return ''
    dict_['entities'] = array
    return (text, dict_)
```

```python
train = [s for s in list(map(lambda s: getEntityTags(s, techNames), sentences[:-5000])) if len(s) > 0 ]
```

```python
ner = nlp.get_pipe('ner')
ner.add_label(tag)
disable_pipes = ['tagger', 'parser']
```

```python
import random
import spacy
from spacy.util import minibatch, compounding
from pathlib import Path
```

```python
print(spacy.__version__)
```

```python
with nlp.disable_pipes(*disable_pipes):
    optimizer = nlp.resume_training()
    for iteration in range(30):
    # shuufling examples  before every iteration
        random.shuffle(train)
        losses = {}
    # batch up the examples using spaCy's minibatch
        batches = minibatch(train, size=compounding(4.0, 32.0, 1.001))
        for batch in batches:
            try:
                texts, annotations = zip(*batch)
                nlp.update(
                        texts,  # batch of texts
                        annotations,  # batch of annotations
                        drop=0.35,  # dropout - make it harder to memorise data
                        losses=losses,
                        sgd = optimizer
                        )
            except:
                print('Ignore')
                continue
            print("Losses", losses)
```

```python
iteration_limit = 100
start = 1.0
end = 4.0
compound = 1.001
dropout = 0.5
with nlp.disable_pipes(*disable_pipes):
    optimizer = nlp.resume_training()
    for iteration in range(iteration_limit):
        
        random.shuffle(train)
        losses = {}
        
        batches = minibatch(train, size = compounding(start, end, compound))
        for batch in batches:
            text, annotations = zip(*batch)
            nlp.update(
                text,
                annotations,
                drop = dropout,
                losses = losses,
                sgd = optimizer
                       )
            print('Loss', losses)
```

```python
result = pd.DataFrame()
old_nlp = en_core_web_sm.load()
description, new, old = [], [], []
for text in sentences[-5000:]:
    doc_updated = nlp(text)
    doc_old = old_nlp(text)
    description.append(text)
    new.append(', '.join([item.text for item in doc_updated.ents]))
    old.append(', '.join([item.text for item in doc_old.ents]))
```

```python
result['description'] = description
result['techstack_entity'] = new
result['original_entity'] = old
result.sample(8)
```
